```python
X = pd.concat(X);
```

```python
data = pd.DataFrame(X);

data.reset_index(inplace = True, drop = True)
for i in range(len(data)):
    data.loc[i,'t'] = i*0.02

# data.set_index('t', inplace=True);
```

```python
# Normalisation
# Feature normalization
# Normalize Time Series Data
# Source : https://machinelearningmastery.com/normalize-standardize-time-series-data-python/

# Normalization is a rescaling of the data from the original range so that all values are within the range of 0 and 1.
# useful when your time series data has input values with differing scales.
features = list(data.columns);
scaledFeatures = [item for item in features if item not in ['t','Gear']]
print(scaledFeatures)
for i in range(len(scaledFeatures)):
    feature = scaledFeatures[i];
    print('Scaling feature : ', scaledFeatures[i])
    values = X[feature].values.reshape(-1,1);
    values = values.astype('float32');
    scaler = MinMaxScaler(feature_range = (0,1));
    scaled_f = scaler.fit_transform(values);
    X[feature] = scaled_f;

```

```python
# Visulaization after normalization

# Histogram
print('Histogram');
data[features].hist(figsize = (20,10));
pyplot.show()

# specify columns to plot
groups = [i for i in range(20)];
values = data.values;
i = 1
# plot each column
print('Plotting each feature(column)');
pyplot.figure(figsize = (15,40))
for group in groups:
	pyplot.subplot(len(groups), 1, i)
	pyplot.plot(values[:, group])
	pyplot.title(data.columns[group], y=0.5, loc='right')
	i += 1
pyplot.show();

```

```python
# Feature importance is one method to help sort out what might be more useful in when modeling.

# load data
array = data.values

# split into input and output
X = array[:,1:-1]
y = array[:,-1]

# fit random forest model
model = RandomForestRegressor(n_estimators=500, random_state=1)
model.fit(X, y)

# show importance scores
print(model.feature_importances_)

```

```python
# plot importance scores
names = data.columns.values[1:-1]
print(names)
ticks = [i for i in range(len(names))]
plt.bar(ticks, model.feature_importances_)
plt.xticks(ticks, names)
plt.show()
```
