## All Gear+Segments Appended (skate_without_video)

```python
# import statements

import pandas as pd
import numpy as np
from matplotlib import pyplot
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from pandas import Series

from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
# Generate dataframe with labeled gear data

# Gear 2
X1_Gear2 = pd.read_csv('./Gear2/segment2.csv', parse_dates=['t'], index_col='t');
features = list(X1_Gear2.columns);
X2_Gear2 = pd.read_csv('./Gear2/segment3.csv', parse_dates=['t'], index_col='t');
X3_Gear2 = pd.read_csv('./Gear2/segment4.csv', parse_dates=['t'], index_col='t');
X4_Gear2 = pd.read_csv('./Gear2/segment5.csv', parse_dates=['t'], index_col='t');
X5_Gear2 = pd.read_csv('./Gear2/segment6.csv', parse_dates=['t'], index_col='t');

X_Gear2 = np.concatenate((X1_Gear2,X2_Gear2,X3_Gear2,X4_Gear2,X5_Gear2), axis=0)
Y_Gear2 = np.full((X_Gear2.shape[0],1), 2)


# Gear 3
X1_Gear3 = pd.read_csv('./Gear3/segment1.csv', parse_dates=['t'], index_col='t');
X2_Gear3 = pd.read_csv('./Gear3/segment2.csv', parse_dates=['t'], index_col='t');
X3_Gear3 = pd.read_csv('./Gear3/segment3.csv', parse_dates=['t'], index_col='t');
X4_Gear3 = pd.read_csv('./Gear3/segment4.csv', parse_dates=['t'], index_col='t');
X5_Gear3 = pd.read_csv('./Gear3/segment5.csv', parse_dates=['t'], index_col='t');
X6_Gear3 = pd.read_csv('./Gear3/segment6.csv', parse_dates=['t'], index_col='t');

X_Gear3 = np.concatenate((X1_Gear3,X2_Gear3,X3_Gear3,X4_Gear3,X5_Gear3,X6_Gear3), axis=0)
Y_Gear3 = np.full((X_Gear3.shape[0],1), 3)


# Gear 4
X1_Gear4 = pd.read_csv('./Gear4/segment2.csv', parse_dates=['t'], index_col='t');
X2_Gear4 = pd.read_csv('./Gear4/segment3.csv', parse_dates=['t'], index_col='t');
X3_Gear4 = pd.read_csv('./Gear4/segment4.csv', parse_dates=['t'], index_col='t');
X4_Gear4 = pd.read_csv('./Gear4/segment5.csv', parse_dates=['t'], index_col='t');
X5_Gear4 = pd.read_csv('./Gear4/segment6a.csv', parse_dates=['t'], index_col='t');
X6_Gear4 = pd.read_csv('./Gear4/segment6b.csv', parse_dates=['t'], index_col='t');

X_Gear4 = np.concatenate((X1_Gear4,X2_Gear4,X3_Gear4,X4_Gear4,X5_Gear4,X6_Gear4), axis=0)
Y_Gear4 = np.full((X_Gear4.shape[0],1), 4)


# Double
X1_Double = pd.read_csv('./Double/segment1.csv', parse_dates=['t'], index_col='t');
X2_Double = pd.read_csv('./Double/segment2.csv', parse_dates=['t'], index_col='t');
X3_Double = pd.read_csv('./Double/segment3.csv', parse_dates=['t'], index_col='t');
X4_Double = pd.read_csv('./Double/segment4.csv', parse_dates=['t'], index_col='t');
X5_Double = pd.read_csv('./Double/segment5.csv', parse_dates=['t'], index_col='t');
X6_Double = pd.read_csv('./Double/segment7.csv', parse_dates=['t'], index_col='t');

X_Double = np.concatenate((X1_Double,X2_Double,X3_Double,X4_Double,X5_Double,X6_Double), axis=0)
Y_Double = np.full((X_Double.shape[0],1), 5)

X = np.concatenate((X_Gear2, X_Gear3, X_Gear4, X_Double), axis = 0);
Y = np.concatenate((Y_Gear2, Y_Gear3, Y_Gear4, Y_Double), axis = 0);

print('Data :', X.shape)
print('Class : ', Y.shape)
```

```python
# Create the dataframes
df = pd.DataFrame(data=X);
Y = pd.DataFrame(data=Y);
```

```python
# Visualization
print('20 Features : ', features)

# Histogram
print('Histogram');
df.hist(figsize = (20,10));
pyplot.show()

# Kernel Density Estimation
print('Kernel Density Estimation');
df.plot(kind='kde')
pyplot.show()

# specify columns to plot
groups = list(df.columns)
values = df.values;
i = 1
# plot each column
print('Plotting each feature(column)');
pyplot.figure(figsize = (15,40))
for group in groups:
	pyplot.subplot(len(groups), 1, i)
	pyplot.plot(values[:, group])
	pyplot.title(df.columns[group], y=0.5, loc='right')
	i += 1
pyplot.show()
```
