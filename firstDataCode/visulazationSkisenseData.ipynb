```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataDemo = pd.read_csv('demo.csv',low_memory = False)
dataDemo.columns = ['time', 'forceLeft','angleLeft','forceRight','angleRight','postionEast','postionsNorth','altitude','speed','freq']
```

```python
dataDemo.head()
```

```python
plt.figure(0,figsize=(15, 6))
plt.plot(dataDemo.time,dataDemo.forceLeft,label='Force [N]')
plt.plot(dataDemo.time,dataDemo.angleLeft,label='Angle [deg]')
plt.plot(dataDemo.time,dataDemo.speed,linewidth=7.0,label='Speed [m/s]')
plt.legend()
plt.title('Time seris of force and angle in the left pole')
plt.figure(1,figsize=(15, 6))
plt.plot(dataDemo.time,dataDemo.forceRight,label='Force [N]')
plt.plot(dataDemo.time,dataDemo.angleRight,label='Angle [deg]')
plt.plot(dataDemo.time,dataDemo.speed,linewidth=7.0,label='Speed [m/s]')
plt.legend()
plt.title('Time seris of force and angle in the right pole')
plt.figure(2,figsize=(15, 6))
plt.plot(dataDemo.time,dataDemo.speed,label='Speed [m/s]')
plt.title('Time series of the speed')
```

The figures above shows timeseries for the original dataset.

In the cell below data is two new dataset is created, nemed dataSkining and dataSkiningForceLeft. The dataset dataSking contains all data from the original dataset where the skier where skining. The other dataset dataSkiningForceLeft contains only data where the force is larger than 1 N. 

```python
skingTimes = [51,122,162,268,334,410,454,544,623,723,757,876,936,1024,1071,1171,1251,1328,1396,1481,1588,1771,1772,2216]
dataSking = dataDemo.loc[(dataDemo['time'] >= 51) & (dataDemo['time'] <= 122)]
for i in range(2,len(skingTimes),2):
    df = dataDemo.loc[(dataDemo['time'] >= skingTimes[i]) & (dataDemo['time'] <= skingTimes[i+1])]
    dataSking = dataSking.append(df)
    
dataSkingForceLeft = dataDemo.loc[dataDemo['forceLeft'] >= 1]
```

```python
timeP = np.linspace(0,2000,dataSking.shape[0])
plt.figure(0,figsize=(15, 6))
plt.plot(timeP,dataSking.speed,label='Speed [m/s]')
plt.title('Time series for speed using dataset dataSking')
plt.legend()
```

In the figure above the time series for the speed in dataset dataSking is ploted.

```python
plt.figure(0)
dataSking.speed.plot(kind='hist', bins=30, alpha=0.7, figsize=[15,6],label = 'Speed')
plt.legend()
plt.title('Speed using dataset dataSking')
plt.figure(1)
dataSking.forceLeft.plot(kind='hist', bins=30, alpha=0.7, figsize=[15,6],label = 'Force left')
plt.legend()
plt.title('Force left using dataset dataSking')
plt.figure(3)
dataSking.forceRight.plot(kind='hist', bins=30, alpha=0.7, figsize=[15,6],label = 'Force right')
plt.legend()
plt.title('Force right using dataset dataSking')
```

The three figures above shows distribution results using dataset dataSking, the three diffrent feature distributions showed are speed, force left, force right respectivtly. 

```python
plt.figure(0)
dataSkingForceLeft.forceLeft.plot(kind='hist', bins=30, alpha=0.7, figsize=[15,6],label='Left pole force')
plt.title('Force left using dataset dataSkingForceLeft')
plt.legend()
```

The figure illustrate the distribution in the left force using dataset dataSkingForceLeft.

```python
plt.figure(0)
dataSkingForceLeft.forceRight.plot(kind='hist', bins=30, alpha=0.7, figsize=[15,6],label='Right pole force')
dataSkingForceLeft.forceLeft.plot(kind='hist', bins=30, alpha=0.7, figsize=[15,6],label='Left pole force')
plt.legend()
plt.title('Force in the left and right pole using dateset dataSkingForceLeft')
```

The figure shows the distribution in the left and the right pole, using data set dataSkingForceLeft. 

```python
plt.figure(0)
dataSking.speed.plot(kind='hist', bins=30, alpha=0.7, figsize=[15,6],label = 'Speed')
plt.legend()
plt.title('Speed')
plt.figure(1)
dataSkingForceLeft.speed.plot(kind='hist', bins=30, alpha=0.7, figsize=[15,6],label='Speed')
plt.legend()
plt.title('Speed')
plt.figure(2)
dataSking.forceLeft.plot(kind='hist', bins=30, alpha=0.7, figsize=[15,6],label = 'Force left')
plt.legend()
plt.title('Force left')
plt.figure(3)
dataSkingForceLeft.forceLeft.plot(kind='hist', bins=30, alpha=0.7, figsize=[15,6],label = 'Force left')
plt.legend()
plt.title('Force left')
plt.figure(4)
dataSking.forceRight.plot(kind='hist', bins=30, alpha=0.7, figsize=[15,6],label = 'Force right')
plt.legend()
plt.title('Force right')
plt.figure(5)
dataSkingForceLeft.forceRight.plot(kind='hist', bins=30, alpha=0.7, figsize=[15,6],label = 'Force right')
plt.legend()
plt.title('Force right')
```
