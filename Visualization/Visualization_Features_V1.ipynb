###### The code below loads the new “demo.csv” file with 9 features and plots each series as a separate subplot

```python
from pandas import read_csv
from matplotlib import pyplot

# load dataset
dataset = read_csv('demo.csv', header=0, index_col=0)
values = dataset.values

# specify columns to plot
groups = [0, 1, 2, 3, 4, 5, 6, 7, 8]
i = 1
# plot each column
pyplot.figure(figsize = (10,8))
for group in groups:
	pyplot.subplot(len(groups), 1, i)
	pyplot.plot(values[:, group])
	pyplot.title(dataset.columns[group], y=0.5, loc='right')
	i += 1
pyplot.show()
```

###### The code below plots the histograms and the kde plots for these 9 features

```python

dataset.hist(figsize = (20,10))
pyplot.show()

dataset.plot(kind='kde')
pyplot.show()
```
